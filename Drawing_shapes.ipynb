```python
# Drew large and small circles using turtle, and practiced filling smaller ones with red and blue colors.
import turtle as t
t.circle(180)
t.color("red")
t.begin_fill()
t.circle(90)
t.end_fill()

t.circle(180)
t.color("blue")
t.begin_fill()
t.circle(90)
t.end_fill()
```

```python
# Drew two filled circles (red and blue) at separate positions using forward and right movement for placement
import turtle as t

t.forward(90)
t.color("red")
t.begin_fill()
t.circle(60)
t.end_fill()


t.forward(90)
t.right(90)
t.forward(90)


t.color("blue")
t.begin_fill()
t.circle(60)
t.end_fill()

```

```python
# Created a colorful rotating flower pattern with 36 overlapping circles using a loop and color cycling
import turtle 
t = turtle.Turtle()
t.speed(0)
colors = ["magenta", "cyan", "lime", "gold"]

for i in range(36):
    t.color(colors[i % len(colors)])
    t.circle(100)
    t.right(10)
```

```python
# Drew a square based on user input for side length using turtle graphics and right turns
import turtle as t

t.reset()
length = int(input("what size of the square do you want?"))

t.forward(length)
t.right(90)
t.forward(length)
t.right(90)
t.forward(length)
t.right(90)
t.forward(length)
t.right(90)
```

```python
# Drew and filled a regular blue hexagon using a loop and turtle's fill functionality.
import turtle as t
t.reset()
length=100
t.color("blue")
t.begin_fill()

for i in range(6):
    t.forward(length)
    t.right(60)
t.end_fill()
    
```

```python
# Loop through numbers starting from 0 up to (but not including) 5,
# and print each number. This demonstrates how Python's `range()` function
# generates a sequence of integers and how a `for` loop can be used
# to iterate through them.
for i in range(5):
    print(f"{i}")
```

```python
# Loop through numbers 0 to 4 using range(5), then add 1 to each value to print numbers 1 to 5
for i in range(5):
    print(f"{i+1}")
```
